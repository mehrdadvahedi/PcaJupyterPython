```python
# ابتدا در اولین بلوک کتابخانه های مورد نیاز را ایمپورت میکنیم.
# کتابخانه برای بازکردن فایل ها
import cv2
# کتابخانه برای برخی اعمال ریاضی
import numpy as np
# کتابخانه برای نمایش تصاویر
import matplotlib.pyplot as plt
# تابع آماده برای کاهش بعد تصاویر 
from sklearn.decomposition import PCA
```

```python
# در ابتدا تصویر را از فایل خوانده و در متغیری ذخیره میکنیم
Picture = cv2.imread("file.jpg")
# حال تصویر ذخیره شده را نمایش میدهیم
plt.imshow(Picture)
plt.show()
```

```python
# همانطور که میبینید تصویر خوانده شده ترکیب رنگی صحیحی ندارد
# فلذا با قطعه کد زیر ترکیب رنگی اشتباه را اصلاح کرده و مجددا نمایش میدهیم
Picture2 = cv2.cvtColor(cv2.imread("file.jpg"), cv2.COLOR_BGR2RGB)
plt.imshow(Picture2)
plt.show()
```

```python
def ourPcaFunc(num):
    global Picture2
    # حال ماتریس های مربوط به رنگ های آبی-سبز-قرمز را از تصویر استخراج میکنیم
    Red, Green, Blue = cv2.split(Picture2)
    # تابع کاهش بعد را با دو پارامتر متفاوت تشکیل میدهیم
    pca1 = PCA(num)
    # حال کاهش بعد را به ماتریس مربوط به رنگ ها اعمال میکنیم
    # قرمز
    RedTransform1 = pca1.fit_transform(Red)
    RedInverse1 = pca1.inverse_transform(RedTransform1)
    
    # سبز
    GreenTransform1 = pca1.fit_transform(Green)
    GreenInverse1 = pca1.inverse_transform(GreenTransform1)
    
    # آبی
    BlueTransform1 = pca1.fit_transform(Blue)
    BlueInverse1 = pca1.inverse_transform(BlueTransform1)
    
    # حال با استفاده از ماتریس های حاصل شده میتوانیم تصویر حاصل را بسازیم
    NpStack1 = np.dstack((RedInverse1, GreenInverse1, BlueInverse1))
    PictureCompress1 = NpStack1.astype(np.uint8)
    
    # حال تصویر را به نمایش در می آوریم
    plt.imshow(PictureCompress1)
    plt.show()
```
